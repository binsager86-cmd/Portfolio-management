```python
import sqlite3
import time
import os

DB_PATH = "portfolio.db"

def test_crud():
    print("Connecting...")
    conn = sqlite3.connect(DB_PATH)
    cur = conn.cursor()
    
    test_user = f"test_user_{int(time.time())}"
    test_email = f"{test_user}@example.com"
    pwd = "password123"
    
    print(f"Registering {test_user}...")
    cur.execute("INSERT INTO users (username, email, password_hash, created_at) VALUES (?, ?, ?, ?)",
               (test_user, test_email, pwd, int(time.time())))
    conn.commit()
    print("Committed.")
    conn.close()
    
    print("Reconnecting to verify...")
    conn2 = sqlite3.connect(DB_PATH)
    cur2 = conn2.cursor()
    cur2.execute("SELECT * FROM users WHERE username=?", (test_user,))
    row = cur2.fetchone()
    print("Result:", row)
    conn2.close()

test_crud()
```
